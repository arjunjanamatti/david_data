```python
import pandas as pd
import numpy as np
```

* Load the excel file to a dataframe

```python
raw_data = pd.read_excel("C:/Users/Arjun Janamatti/Downloads/Python question final.xlsx", index_col=[0,1,2], header=[0,1,2])
raw_data
```

* Access vertical zones

```python
raw_data_vertical = raw_data.xs('province',axis=1, level=1, drop_level=False)
vertical_zones = (raw_data_vertical.xs('Jing',axis=1, level=2, drop_level=False))
vertical_zones
```

* Splitting the vertical zone values based on their country

```python
zones_vertical = []
rows_list_level_0 = (vertical_zones.index).get_level_values(0).unique()
for i in range(len(rows_list_level_0)):
    if (vertical_zones.index).get_level_values(0).unique()[i] == rows_list_level_0[i]:
        zones_vertical.append((vertical_zones.loc[[rows_list_level_0[i]]]).values)
zones_vertical
```

* Adding the co-effecients

```python
vertical_zone_coeffecients = [[0.25, 0.25, 0.5],[0.15, 0.15, 0.7],[0.1, 0.1, 0.8]]
```

* Perform calculations
* The multiplication is not element wise, hence we will again have to arrange this matrix

```python
new_vertical_values = []
new_list_vertical = []
for i in range(len(zones_vertical)):
    for k in range(len(vertical_zone_coeffecients[i])):
        new_list_vertical.append(zones_vertical[i] * vertical_zone_coeffecients[i][k])
#     print((np.array(new_list)))
    new_vertical_values.extend(new_list_vertical)
    new_list_vertical.clear()
new_vertical_values
```

* Rearrangement of the array as per our shape requirement

```python
vertical_zones_1 = []
for i in range(len(vertical_zone_coeffecients)):
    if i+6 < len(new_vertical_values):
        vertical_zones_1.append(np.hstack(np.r_[new_vertical_values[i],new_vertical_values[i+3], new_vertical_values[i+6]]))
vertical_zones_1
```

* Adding the results to the dataframe, I will copy the original dataframe to another name for simplicity

```python
raw_data_vertical = raw_data.copy(deep=True)
for i in range(len(vertical_zones_1)):
    raw_data_vertical["China","province","level_2_{}".format(i)] = vertical_zones_1[i]
raw_data_vertical
```

#### Accessing the horizontal rows

```python
horizontal_zones = (raw_data_1.xs('Jing',axis=0, level=2, drop_level=False))
horizontal_zones_array = horizontal_zones.values
horizontal_zones_array
```

#### Extract the values for each country

```python
zones = []
columns_list_level_0 = (horizontal_zones.columns).get_level_values(0).unique()
for i in range(len(columns_list_level_0)):
    if (horizontal_zones.columns).get_level_values(0).unique()[i] == columns_list_level_0[i]:
        zones.append((horizontal_zones[[columns_list_level_0[i]]]).values)
zones
```

```python
zone_coeffecients = [[0.2, 0.4, 0.4],[0.3, 0.3, 0.4],[0.1, 0.1, 0.8]]
```

Perform calculations

```python
new_horizontal_values = []
new_list = []
for i in range(len(zones)):
    for k in range(len(zone_coeffecients)):
        new_list.append(zones[i] * zone_coeffecients[i][k])
#     print((np.array(new_list)))
    new_horizontal_values.extend(new_list)
    new_list.clear()
new_horizontal_values
```

#### Arrange the values in form of rows to be added to the main data

```python
horizontal_zones_1 = []
for i in range(len(zone_coeffecients)):
    if i+6 < len(new_horizontal_values):
        horizontal_zones_1.append(np.hstack(np.c_[new_horizontal_values[i],new_horizontal_values[i+3], new_horizontal_values[i+6]]))
horizontal_zones_1
```

```python
raw_data_horizontal = raw_data.copy(deep=True)
for i in range(len(horizontal_zones_1)):
    row_to_append = pd.Series(horizontal_zones_1[i])
#     if i + 1 < len(new_horizontal_values_1):
    raw_data_horizontal.loc[((horizontal_zones.index.get_level_values(0).unique())[0], (horizontal_zones.index.get_level_values(1).unique())[0], "new_row_1_1_{}".format(i))] = row_to_append.values
raw_data_horizontal
```

```python
option_select = input("Enter either '1' for selecting option_1 or '2' for selectin option_2: ")
if option_select == "1":
    print("option_1 is selected")
    final_option_coeffecient = [[0.15,0.15,0.7],[0.15,0.15,0.7],[0.15,0.15,0.7]]
elif option_select == "2":
    print("option_2 is selected")
    final_option_coeffecient = [[1,0,0],[0,1,0],[0,0,1]]
```

Extract the value for calculations

```python
final_option = (raw_data_horizontal.xs('China',axis=1, level=0, drop_level=False))
final_option = (final_option["China","province","Jing"]).values
final_option = final_option[-3:]
final_option
```

Perform the calculations

```python
new_final_option_values = []
new_list_final_option = []
for i in range(len(final_option)):
    for k in range(len(final_option_coeffecient)):
        new_list_final_option.append(final_option[i] * final_option_coeffecient[i][k])
#     print((np.array(new_list_final_option)))
    new_final_option_values.extend(new_list_final_option)
    new_list_final_option.clear()
new_final_option_values = np.split(np.array(new_final_option_values),len(final_option))
new_final_option_values
```

```python
horizontal_zones_2 = []
for i in range(len(horizontal_zones_1)):
    horizontal_zones_2.append(np.r_[horizontal_zones_1[i],new_final_option_values[i]])
horizontal_zones_2
```
